```python
numeric_df = df.select_dtypes(include=[np.number])
numeric_df
```

```python
# To remove outliers
Q1=numeric_df.quantile(0.25)
Q3=numeric_df.quantile(0.75)
IQR=Q3-Q1
ub=Q3+1.5*IQR
lb=Q1-1.5*IQR
```

```python
print(Q1)
print('****************************')
print(Q3)
print('****************************')
print(IQR)
print('****************************')
print(ub)
print('****************************')
print(lb)
```

```python
outlier=numeric_df[((numeric_df<lb)|(numeric_df>ub)).any(axis=1)]
print(outlier)
```

```python
df1=numeric_df[~((numeric_df<lb)|(numeric_df>ub)).any(axis=1)]
```

```python
X=df1.drop('Profit',axis=1) #independenet variables
Y=df1['Profit'] # dep var
```

```python
X.head()
```

```python
Y.head()
```

```python
####Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
## bilding a linear regression model
from sklearn.linear_model import LinearRegression 
lr=LinearRegression()
lr.fit(X_train,Y_train) 
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
Y_pred=lr.predict(X_test)
Y_pred
```

```python
pred=pd.DataFrame()
pred['Actual']=Y_test
pred['predicted']=Y_pred
pred
```

```python
## to find accuracy of model
## R2 score
from sklearn.metrics import r2_score
acuuracy=r2_score(Y_test,Y_pred)
acuuracy
```

```python
#building a OlS model
from statsmodels.regression.linear_model import OLS
import statsmodels.regression.linear_model as srf
```

```python
ols_model=srf.OLS(endog=Y_train,exog=X_train).fit()
ols_model.summary()
```
